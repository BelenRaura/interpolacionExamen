```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

# Definir los puntos
x = np.array([-1, 1])
y = np.array([1, 3])

# Definir las derivadas en los extremos
bc_type = ((1, 1.0), (1, 2.0))

# Crear el spline cúbico
cs = CubicSpline(x, y, bc_type=bc_type)

# Crear un rango de valores x para evaluar el spline
x_new = np.linspace(-1, 1, 100)
y_new = cs(x_new)

# Graficar el spline
plt.plot(x, y, 'o', label='Puntos dados')
plt.plot(x_new, y_new, label='Spline cúbico')

# Mostrar la función spline cúbico
print("La función spline cúbico es:")
print(cs)
def spline_function(x):
    return cs(x)

# Imprimir la función spline cúbico en forma de ecuaciones
coefs = cs.c
print("Coeficientes del spline cúbico por tramos:")
for i in range(len(coefs[0])):
    print(f"Tramo {i+1}:")
    print(f"{coefs[3, i]} + {coefs[2, i]}*(x - {x[i]}) + {coefs[1, i]}*(x - {x[i]})^2 + {coefs[0, i]}*(x - {x[i]})^3")

# Ejemplo de uso de la función spline
x_example = -0.5
y_example = spline_function(x_example)
print(f"f({x_example}) = {y_example}")

# Ejemplo de uso de la función spline
x_example = 0.5
y_example = spline_function(x_example)
print(f"f({x_example}) = {y_example}")

plt.legend()
plt.show()
```

```python
# Definir los nuevos puntos
x_new_points = np.array([-1, 0, 1])
y_new_points = np.array([1, 5, 3])

# Crear el spline cúbico para los nuevos puntos
cs_new = CubicSpline(x_new_points, y_new_points)

# Crear un rango de valores x para evaluar el nuevo spline
x_new_range = np.linspace(-1, 1, 100)
y_new_range = cs_new(x_new_range)

# Graficar el nuevo spline
plt.plot(x_new_points, y_new_points, 'o', label='Puntos dados')
plt.plot(x_new_range, y_new_range, label='Spline cúbico')

# Mostrar la función spline cúbico
print("La función spline cúbico es:")
print(cs_new)

# Imprimir la función spline cúbico en forma de ecuaciones
coefs_new = cs_new.c
print("Coeficientes del spline cúbico por tramos:")
for i in range(len(coefs_new[0])):
    print(f"Tramo {i+1}:")
    print(f"{coefs_new[3, i]} + {coefs_new[2, i]}*(x - {x_new_points[i]}) + {coefs_new[1, i]}*(x - {x_new_points[i]})^2 + {coefs_new[0, i]}*(x - {x_new_points[i]})^3")

plt.legend()
plt.show()
```

* Grafique los splines cúbicos tal que cumplan con una pendiente de $m=2$ en los puntos dados.

```python
# Definir los puntos
x_points = np.array([-1, 0, 1])
y_points = np.array([1, 5, 3])

# Definir las derivadas en los extremos (pendiente m=2)
bc_type = ((1, 2.0), (1, 2.0))

# Crear el spline cúbico con las nuevas condiciones de contorno
cs_with_slope = CubicSpline(x_points, y_points, bc_type=bc_type)

# Crear un rango de valores x para evaluar el spline
x_range = np.linspace(-1, 1, 100)
y_range = cs_with_slope(x_range)

# Graficar el spline
plt.plot(x_points, y_points, 'o', label='Puntos dados')
plt.plot(x_range, y_range, label='Spline cúbico con pendiente m=2')

# Mostrar la función spline cúbico
print("La función spline cúbico con pendiente m=2 es:")
print(cs_with_slope)

# Imprimir la función spline cúbico en forma de ecuaciones
coefs_slope = cs_with_slope.c
print("Coeficientes del spline cúbico por tramos:")
for i in range(len(coefs_slope[0])):
    print(f"Tramo {i+1}:")
    print(f"{coefs_slope[3, i]} + {coefs_slope[2, i]}*(x - {x_points[i]}) + {coefs_slope[1, i]}*(x - {x_points[i]})^2 + {coefs_slope[0, i]}*(x - {x_points[i]})^3")

plt.legend()
plt.show()
```
